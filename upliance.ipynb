```python
import pandas as pd
import numpy as np
```

```python
excel_file = pd.ExcelFile("Assignment.xlsx")
sheet_names = excel_file.sheet_names

dataframes = {}
for sheet_name in sheet_names:
    dataframes[sheet_name] = excel_file.parse(sheet_name)
```

```python
UserDetails = dataframes['UserDetails.csv']
CookingSessions = dataframes['CookingSessions.csv']
OrderDetails = dataframes['OrderDetails.csv']
```

```python
OrderDetails['Rating'] = OrderDetails['Rating'].fillna(0)
```

```python
OrderDetails.isnull().sum()
```

```python
#Merge datasets based on common columns UserID
Cooking_Session_details = UserDetails.merge(CookingSessions, on='User ID')
Food_Order_Details = UserDetails.merge(OrderDetails, on='User ID')
```

```python
Cooking_Session_details
```

```python
Food_Order_Details
```

```python
#Total sessions per user
session_counts = Cooking_Session_details.groupby('User ID')['Session ID'].count().reset_index()
session_counts.columns = ['User ID', 'TotalSessions']
```

```python
session_counts
```

```python
#Average session duration
average_duration = Cooking_Session_details.groupby('User ID')['Duration (mins)'].mean().reset_index()
average_duration.columns = ['User ID', 'AverageDuration']
```

```python
average_duration
```

```python
#Most popular dishes
popular_dishes = Cooking_Session_details['Dish Name'].value_counts()
popular_dishes
```

```python
#Dish preference by age group
dish_by_age = Cooking_Session_details.groupby('Age')['Dish Name'].value_counts()
dish_by_age
```

```python
import matplotlib.pyplot as plt

#Group the data by 'Age' and 'DishName', and count the orders
age_dish_counts = Food_Order_Details.groupby(['Age', 'Dish Name'])['Order ID'].count().reset_index()

#Create a pivot table to restructure the data
pivot_table = age_dish_counts.pivot_table(index='Age', columns='Dish Name', values='Order ID', fill_value=0)

pivot_table.plot(kind='bar', figsize=(15, 10))
plt.title('Dish Preference by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#Analyze cooking sessions and order relationship
session_order_relation = Food_Order_Details.groupby(['Session ID', 'Order ID']).size().reset_index(name='Count')
session_order_relation
```

```python
#Total order amount per user
total_order_amount = Food_Order_Details.groupby('User ID')['Amount (USD)'].sum().reset_index()
total_order_amount.columns = ['User ID', 'TotalOrderAmount']
total_order_amount
```

```python
#Order frequency over time
order_frequency = Food_Order_Details.groupby('Order Date')['Order ID'].count().reset_index()
order_frequency
```

```python
#Order preference by location
order_by_location = Food_Order_Details.groupby('Location')['Dish Name'].value_counts()
order_by_location
```

```python
import matplotlib.pyplot as plt

# Group the data by 'Location' and 'DishName', and count the orders
location_dish_counts = Food_Order_Details.groupby(['Location', 'Dish Name'])['Order ID'].count().reset_index()

# Create a pivot table to restructure the data
pivot_table = location_dish_counts.pivot_table(index='Location', columns='Dish Name', values='Order ID', fill_value=0)

# Select top 5 locations with the highest number of orders
top_locations = pivot_table.sum(axis=1).nlargest(5).index

for location in top_locations:
    plt.figure(figsize=(6, 4))
    pivot_table.loc[location].plot(kind='pie', autopct='%1.1f%%')
    plt.title(f'Dish Preference in {location}')
    plt.show()
```

```python
#Common Users in Cooking Sessions and Order Details
common_users = set(Cooking_Session_details['User ID']).intersection(set(Food_Order_Details['User ID']))
common_users
```

```python
#Dish Preference Comparison
popular_cooked_dishes = Cooking_Session_details['Dish Name'].value_counts().head(10)
popular_ordered_dishes = Food_Order_Details['Dish Name'].value_counts().head(10)
```

```python
popular_cooked_dishes
```

```python
popular_ordered_dishes
```

```python
import matplotlib.pyplot as plt

#Combine the two Series into a DataFrame
comparison_df = pd.concat([popular_cooked_dishes, popular_ordered_dishes], axis=1, keys=['Cooked', 'Ordered'])

comparison_df.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Popular Cooked and Ordered Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#Analyze time-based patterns
time_based_orders = Food_Order_Details.groupby('Time of Day')['Order ID'].count().reset_index()
time_based_orders.columns = ['Time of Day', 'Order Count']
time_based_orders
```

```python
import matplotlib.pyplot as plt
time_based_orders.plot(kind='bar', x='Time of Day', y='Order Count')
plt.xlabel('Time Period')
plt.ylabel('Number of Orders')
plt.title('Order Distribution by Time Period')
plt.show()
```

```python
import matplotlib.pyplot as plt

#Calculate the average rating for each dish in both DataFrames
avg_cooking_ratings = Cooking_Session_details.groupby('Dish Name')['Session Rating'].mean().sort_values(ascending=False)
avg_order_ratings = Food_Order_Details.groupby('Dish Name')['Rating'].mean().sort_values(ascending=False)

#Combine the two Series into a DataFrame
comparison_df = pd.concat([avg_cooking_ratings, avg_order_ratings], axis=1, keys=['Avg Cooking Rating', 'Avg Order Rating'])

comparison_df.plot(kind='bar', figsize=(10, 6))
plt.title('Average Ratings for Cooked and Ordered Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group the data by 'MealType' and count the number of orders
meal_wise_orders = Food_Order_Details.groupby('Meal Type')['Order ID'].count().reset_index()

# Plot the results
plt.figure(figsize=(8, 6))
plt.bar(meal_wise_orders['Meal Type'], meal_wise_orders['Order ID'])
plt.xlabel('Meal Type')
plt.ylabel('Number of Orders')
plt.title('Dish Orders by Meal Type')
plt.show()
```
